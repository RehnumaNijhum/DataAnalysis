```python
#checking ratings column
Dataset.ratings.head()
```

```python
#importing abstract sentex tree for converting the series str to list
import ast
```

```python
#converting the series to list
ast.literal_eval(Dataset.ratings[0])
```

```python
#checking the type
type(ast.literal_eval(Dataset.ratings[0]))
```

```python
#creating another series to store unpack list column
Dataset['ratings_list'] = Dataset.ratings.apply(lambda x: ast.literal_eval(x) )
```

```python
Dataset.head()
```

```python
#checking the unpack data
Dataset.ratings_list[0]
```

```python
#7. Count the total number of ratings received by each talk
```

```python
# creating function to get count coumn from list
def number_of_ratings(dics):
    num = 0
    for i in dics:
        num = num+i['count']
    return num
```

```python
number_of_ratings(Dataset.ratings_list[0])
```

```python
#getting all total
Dataset.ratings_list.apply(number_of_ratings).head()
```

```python
# creating coulmn for total ratings
Dataset['total_ratings'] = Dataset.ratings_list.apply(number_of_ratings)
```
