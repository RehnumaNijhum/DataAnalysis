```python
import pandas as pd
```

```python
DataSet = pd.read_csv('winequality-red.csv')
```

```python
DataSet.head()
```

```python
# rename column name in uppercase/lowercase
DataSet.columns = [x.lower() for x in DataSet.columns]
```

```python
DataSet
```

```python
# remove space between column name
DataSet.columns = DataSet.columns.str.replace(' ', '_')
```

```python
DataSet
```

```python
# rename column name 
DataSet.rename(columns = {'ph': 'PH'}, inplace = True)
```

```python
DataSet
```

```python
#drop column
DataSet.drop(columns=['volatile_acidity'], inplace = True)
```

```python
DataSet
```

```python
#add values in a column
DataSet.append({'column_name':'value'}, ignore_index = True)
```

```python
# removing rows
#use inplace=True for parmanently remove from dataframe
DataSet.drop(DataSet.index[1:4])
```

```python
#removing rows by condition
Filter = DataSet['quality'] <5
DataSet.drop(index=DataSet[Filter].index)
```

```python
#sorting values
DataSet.sort_values(by = ['alcohol','quality'],ascending= False)
```
